**Resumo:** O trabalho apresentado em ***phyton*** pretende resolver o problema proposto para TPC relativamente a interfaces. Neste trabalho foi criada uma interface para o programa criado na aula para o processamento do ***Dataset de exames médicos***.

```python
import matplotlib.pyplot as plt
import PySimpleGUI as sg


def getEMD(linha):
    emd=[]
    novaLinha=linha.replace("\n","")
    campos=novaLinha.split(",")
    emd.append("emd"+str(int(campos[1])+1))
    for i in range(2,len(campos)):
            emd.append(campos[i])
    return emd

def lerDataset(fnome):
    f = open(fnome, encoding="utf-8")
    bd = []
    f.readline() #retirar a linha de cabeçalho
    for linha in f:
        emd=getEMD(linha)
        bd.append(emd)
    return bd

def chaveOrd(a):
    data=a[1]
    data=data.split("-")
    return data

def listarDataset(bd):
    bd.sort(key=chaveOrd,reverse=True)
    print("id           Data           Nome        Apto")
    print("----------------------------------------------")
    for i in bd:
        print(i[0]+ " | "+i[1]+ " | "+i[3]+" "+i[2]+" | "+i[11])
    return

def consultarDataset(bd, id):
    i=0
    while(bd[i][0]!=id):
        i=i+1
    return bd[i]


def chaveOrd2(a):
    return a[7]

def modalidades(bd):
    bd.sort(key=chaveOrd2)
    L=[]
    for emd in bd:
        if(emd[7] not in L):
            L.append(emd[7])
    return L

def distribPorModalidade(bd):
    modalidade={}
    for a in bd:
        if a[7] in modalidade.keys():
            modalidade[a[7]]=modalidade[a[7]]+1
        else:
            modalidade[a[7]]=1
    return modalidade

def distribPorClube(bd):
    clube={}
    for a in bd:
        if a[8] in clube.keys():
            clube[a[8]]=clube[a[8]]+1
        else:
            clube[a[8]]=1
    return clube

def distribPorAno(bd):
    ano={}
    for a in bd:
        data=a[1]
        data=data.split("-")
        if data[0] in ano.keys():
            ano[data[0]]=ano[data[0]]+1
        else:
            ano[data[0]]=1
    return ano

def distrib(bd,i):
    modalidade={}
    for a in bd:
        if a[i] in modalidade.keys():
            modalidade[a[i]]=modalidade[a[i]]+1
        else:
            modalidade[a[i]]=1
    return modalidade

def plotDistribPorModalidade(bd):
    d=distribPorModalidade(bd)
    x=d.keys()
    y=[]
    for i in x:
        y.append(d[i])
    plt.xlabel("Modalidade")
    plt.ylabel("Nº de alunos")
    plt.title("Distribuição por modalidades")
    plt.bar(x,y,align='center', width=0.8)
    plt.show()
    
def plotDistrib(bd):
    d=distrib(bd,4)
    x=d.keys()
    y=[]
    for i in x:
        y.append(d[i])
    plt.title("Distribuição")
    plt.bar(x,y)
    plt.show()



#Interface           
menu_list_column = [
    [sg.Button("Carregar BD")],
    [sg.Button("Modalidades")],
    [sg.Button("Destribuição por Ano")],
    [sg.Button("Destribuição por Clube")],
    [sg.Button("Destribuição por Modalidades")],
    [sg.Button("Gráfico da destribuição por Modalidades")],
    [sg.Button("Sair")]
]

data_viewer_column = [
    [sg.Text("Painel de Dados")],
    [sg.Text(size=(50, 10), key="-Dados-")],
]

# ----- Full layout -----
layout = [
    [
        sg.Column(menu_list_column),
        sg.VSeperator(),
        sg.Column(data_viewer_column),
    ]
]

window = sg.Window("Operações BD", layout,  margins=(100, 100))
# Run the Event Loop
stop = False
while not stop:
    event, values = window.read()
    if event == "Sair" or event == sg.WIN_CLOSED:
        stop = True
        
    elif event == "Carregar BD":
        window["-Dados-"].update(lerDataset('emd.csv'))
    
    elif event == "Modalidades":
        window["-Dados-"].update(modalidades(lerDataset('emd.csv')))

    elif event == "Destribuição por Ano":  
        window["-Dados-"].update(distribPorAno(lerDataset('emd.csv')))
        
    elif event == "Destribuição por Clube":  
        window["-Dados-"].update(distribPorClube(lerDataset('emd.csv')))
        
    elif event == "Destribuição por Modalidades":  
        window["-Dados-"].update(distribPorModalidade(lerDataset('emd.csv')))
 
window.close()
```
