```python
import matplotlib.pyplot as plt
import PySimpleGUI as sg

#Exercício 1:

def getAluno(texto):
    textoLimpo2=texto.replace("\"","")
    campos=textoLimpo2.split(",")
    aluno=[]
    aluno.append(campos[0])
    aluno.append(campos[1])
    aluno.append(campos[2])
    aluno.append(campos[3:])
    lista=[]
    for s in aluno[3]:
        s=lista.append(int(s))
    aluno[3]=lista
    return aluno


def lerDataset(fnome):
    f=open(fnome)
    bd=[]
    f.readline()
    for linha in f:
        bd.append(getAluno(linha))
    return bd

#Exercício 2:

def chaveOrd(a):
    return a[1]

def listarDataset(bd):
    bd.sort(key=chaveOrd)
    print(" id   nome  curso   média")
    print("-------------------------")
    for a in bd:
        print(a[0]+"|"+a[1]+"|"+a[2]+"|"+str(sum(a[3])/4))



#Exercício 3:

def consultarDataset(bd, id):
    for a in bd:
        for a[0] in a:
            if id == a[0]:
                print((a[0])+"|"+(a[1])+"|"+(a[2])+"|"+str(sum(a[3])/4))
            else:
                print("Não há nenhum aluno com esse id!")

#Exercício 4:
def media(l):
    soma=0
    for i in range(0,len(l)):
        soma=soma+l[i]
    return soma/4


def chaveOrdena(a):
    return media(a[3])

def top10(bd):
    bd.sort(key=chaveOrdena, reverse = True)
    top10=[]
    for a in range (0, 10):
        aluno = []
        for i in range(0,3):
            aluno.append(bd[a][i])
        top10.append(aluno)
    return top10


#Exercício 5:

def distribPorCurso(bd):
    distribuicao={}
    for a in bd:
        if a[2] in distribuicao.keys():
            distribuicao[a[2]]= distribuicao[a[2]]+1
        else:
            distribuicao[a[2]]= 1
    return distribuicao


#Exercício 6:

def distribPorMedia(bd):
    distribuicao={}
    for a in bd:
        if ("Média=", (sum(a[3])/4)) in distribuicao.keys():
            distribuicao[("Média=", (sum(a[3])/4))] = distribuicao[("Média=", (sum(a[3])/4))]+1
        else:
            distribuicao[("Média=", (sum(a[3])/4))] = 1
           
    return distribuicao


#Exercício 7:

def plotDistribPorCurso(bd):
    curso = distribPorCurso(bd)
    x = curso.keys()
    y = []
    for n in x:
        y.append(curso[n])
    plt.bar(x, y, width = 0.8, color='purple')
    plt.xlabel('Curso')
    plt.ylabel('nº de alunos')
    plt.title('Distribuição por curso')
    plt.show()


#Exercício 8:

def plotDitribPorMedia(bd):
    media = distribPorMedia(bd)
    x = media.keys()
    y = []
    for n in x:
        y.append(media[n])
    plt.bar(x, y, color='purple')
    plt.xlabel('Média')
    plt.title('Distribuição por médias')
    plt.show()

    
BD = []
BD = lerDataset("alunos-windows.csv")
print("EXERCÍCIO 2: ")
print("\n")
print(listarDataset(BD))
print("EXERCÍCIO 4: ")
print("\n")
for linha in top10(BD):
    print(linha)
print("\n")
print("EXERCÍCIO 5: ")
print("\n")
print(distribPorCurso(BD))
print("\n")
print("EXERCÍCIO 6: ")
print("\n")
print(distribPorMedia(BD))
print("\n")
print("EXERCÍCIO 7: ")
print("\n")
plotDistribPorCurso(BD)
print("\n")
print("EXERCÍCIO 8: ")
print("\n")
plotDitribPorMedia(BD)    
```
